```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 200)
```

```python
df = pd.read_csv("../data/processed/anime_cleaned.csv")
```

```python
features_df = df[
    [
        "mal_id",
        "title",
        "title_english",
        "genres",
        "themes",
        "studios",
        "demographics",
        "type",
        "synopsis",
        "url"
    ]
].copy()
```

```python
features_df.head()
```

```python
features_df.info()
```

```python
text_columns = [
    "genres",
    "themes",
    "studios",
    "demographics",
    "type"
]


for col in text_columns:
    features_df[col] = (
        features_df[col]
        .str.lower()
        .str.replace(",", " ", regex=False)
        .str.replace(" ", "_", regex=False)
    )
```

```python
features_df["synopsis"] = (
    features_df["synopsis"]
    .fillna("")
    .str.lower()
)
```

```python
features_df.head()
```

```python
boilerplate = (
    "No synopsis has been added for this series yet.\n\n"
    "Click here to update this information."
)

features_df["synopsis"] = (
    features_df["synopsis"].replace(boilerplate, "")
)
```

```python
features_df["combined_features"] = (
    features_df["genres"] + " " +
    features_df["themes"] + " " +
    features_df["studios"] + " " +
    features_df["demographics"] + " " +
    features_df["type"] + " " +
    features_df["synopsis"]
)
```

```python
features_df[
    ["title", "combined_features"]
].head()
```

```python
recommendation_df = features_df[
    [
        "mal_id",
        "title",
        "combined_features",
        "url"
    ]
].copy()
```

```python
recommendation_df.info()
```

```python
recommendation_df.to_csv(
    "../data/processed/anime_recommendation_features.csv",
    index=False
)
```
