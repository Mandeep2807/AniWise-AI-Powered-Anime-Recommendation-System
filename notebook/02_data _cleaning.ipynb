### Removing Duplicate Records

Duplicate records were identified and removed to ensure data consistency and prevent redundant information from affecting analysis and the recommendation system.

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": round((df.isnull().sum() / len(df)) * 100, 2)
})

missing = missing[missing["Missing Values"] > 0].sort_values(
    by="Percentage",
    ascending=False
)

missing
```

```python
df["themes"] = df["themes"].fillna("Unknown")
df["studios"] = df["studios"].fillna("Unknown")
df["demographics"] = df["demographics"].fillna("Unknown")
df["genres"] = df["genres"].fillna("Unknown")
```

```python
df["synopsis"] = df["synopsis"].fillna("")
```

```python
df["type"] = df["type"].fillna(df["type"].mode()[0])
```

```python
df["episodes"] = df["episodes"].fillna(df["episodes"].median())
```
